```python
import numpy as np
from config import *
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
train_labels = np.load(data_dir + 'train_labels.npy')
train_labels = train_labels[train_labels > 0]
```

```python
gamma = stats.gamma.fit(train_labels)
log_normal = stats.lognorm.fit(train_labels, floc=0)
```

```python
plt.hist(train_labels, bins=40, density=True)
x = np.linspace(0, 1000, 10000)
plt.plot(x, stats.gamma.pdf(x, *gamma), 'r-', lw=2, label='gamma pdf')
plt.plot(x, stats.lognorm.pdf(x, *log_normal), 'g-', lw=2, label='log-normal pdf')
plt.legend(loc='upper right')
plt.show()
```
