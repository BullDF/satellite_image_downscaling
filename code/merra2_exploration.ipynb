```python
import pandas as pd
import numpy as np
import netCDF4 as nc
import xarray as xr
import re
import pyreadr
```

```python
data_dir = '../data/'
```

```python
merra2_2000_2018 = xr.open_dataset(data_dir + 'MERRA2_aerosol_variables_over_MiddleEast_daily_20000516-20180515.nc')
```

```python
merra2_2000_2018
```

```python
# Filtrer les données après 2017
merra2_2017 = merra2_2000_2018.sel(time=slice('2017-01-01', '2017-12-31'))
```

```python
merra2_2017
```

```python
merra2_2018_2020 = xr.open_dataset(data_dir + 'MERRA2_aerosol_variables_over_MiddleEast_daily_20180101-20200531.nc')
```

```python
merra2_2018_2020
```

```python
merra2_2017_2020 = xr.concat([merra2_2017, merra2_2018_2020], dim='time').pipe(
    lambda ds: ds.assign(
        lat=ds.lat.isel(time=0),
        lon=ds.lon.isel(time=0)
    )
)
```

```python
merra2_2017_2020
```

```python
merra2_2017_2020.to_netcdf(data_dir + "merra2-2017-2020.nc")
```
