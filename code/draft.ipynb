```python
files = os.listdir(surface_flux_dir)
```

```python
for f in tqdm(files):
    if f.endswith('.nc4'):
        date = f[:8]
        ds = xr.open_dataset(surface_flux_dir + f)
        assert len(ds['time'].values == 24)
        for time in ds['time'].values:
            t = np.datetime_as_string(time)
            assert t[0:4] + t[5:7] + t[8:10] == date
```
