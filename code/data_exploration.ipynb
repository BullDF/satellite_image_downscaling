```python
import re
from datetime import datetime

# Expressions régulières pour les deux formats
utc_pattern = r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})T(?P<hour>\d{2}):(?P<minute>\d{2}):(?P<second>\d{2}\.\d{3})Z'
local_pattern = r'(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})T(?P<hour>\d{2}):(?P<minute>\d{2}):(?P<second>\d{2})\+(?P<offset>\d{2}):(?P<offset_minute>\d{2})'

# Example strings
utc_string = '2017-08-11T05:00:00.000Z'
local_string = '2017-08-11T08:00:00+03:00'

# Function to parse datetime with regex
def parse_datetime_with_regex(date_string, pattern):
    match = re.match(pattern, date_string)
    if match:
        parts = match.groupdict()
        return datetime(
            int(parts['year']),
            int(parts['month']),
            int(parts['day']),
            int(parts['hour']),
            int(parts['minute']),
            int(float(parts['second'])),  # handling seconds with decimals
            int((float(parts['second']) - int(float(parts['second']))) * 1_000_000)  # microseconds
        )
    return None

# Parsing the UTC and local datetime strings
utc_datetime = parse_datetime_with_regex(utc_string, utc_pattern)
local_datetime = parse_datetime_with_regex(local_string, local_pattern)

print(f"UTC datetime: {utc_datetime}")
print(f"Local datetime: {local_datetime}")
```
